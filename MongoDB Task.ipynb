```python
import pymongo
import pandas as pd
import json
import logging as lg
import os

#Log File
lg.basicConfig(filename = 'mongotest.log', level = lg.INFO, filemode = 'a', format = '%(asctime)s - %(levelname)s - %(message)s')


#File reading and Data cleaning
try:
    df = pd.read_csv('carbon_nanotubes.csv',sep=';')
    try:
        df['Initial atomic coordinate u'] = df['Initial atomic coordinate u'].str.split(',')
        df['Initial atomic coordinate v'] = df['Initial atomic coordinate v'].str.split(',')
        df['Initial atomic coordinate w'] = df['Initial atomic coordinate w'].str.split(',')
        df["Calculated atomic coordinates u'"] = df["Calculated atomic coordinates u'"].str.split(',')
        df["Calculated atomic coordinates v'"] = df["Calculated atomic coordinates v'"].str.split(',')
        df["Calculated atomic coordinates w'"] = df["Calculated atomic coordinates w'"].str.split(',')
        lg.info("File Loaded Successfully and Data is cleaned")
    except:
        lg.info('Data cleaning not possible')  
        
    #Python-Mongodb Connection establish
    client = pymongo.MongoClient("mongodb+srv://mongodb:mongodb@cluster0.akfky.mongodb.net/myFirstDatabase?retryWrites=true&w=majority")
    db = client.test

    #Database Created or Selected
    db = client['harsh']

    #Collection Created or Selected
    coll2 = db['MongoTest']

    #converting csv file into json file
    data = df.to_dict(orient='records')
        
except Exception as e:
    lg.info(e)
            
```

```python
class connection(Exception):
        def insert_operation():
            """This function will insert the data into Mongodb"""
            try:
                coll2.insert_many(data)
                lg.info('All data are inserted into Mongodb Atlas!')
            except:
                lg.info('Data is not inserted into Mongodb Atlas')

        
        def find_operation(): 
            """This function will find the data based on condition"""
            try: 
                for i in coll2.find({'Initial atomic coordinate u': ['0', '679005']}).limit(2):
                    lg.info('Find Successfull!')
                    continue
                raise Exception('Not found anything')
            except Exception as e:
                lg.exception(e)

    
        def filter_operation(): 
            """This function will filter the data based on condition"""
            try:
                for i in coll2.find({'Chiral indice m':{'$gte': 6}}).limit(2):
                    lg.info('Filter Successfull!')
                    continue
                raise Exception('Filter Error Exception')
            except Exception as e:
                lg.exception(e)


        def update_operation():
            """This function will update the data based on condition"""
            try:
                u = {'Chiral indice n': 4}
                f = coll2.find(u)
                for i in f:
                    coll2.update_many(u,{'$set':{'Chiral indice n': 40}})
                    lg.info('Record Updated!')
                    continue
                raise Exception('Record Updation Error')
            except Exception as e:
                lg.exception(e)


        def delete_operation():
            """This function will delete the data based on condition"""
            try:
                u = {'Chiral indice n': 5}
                f = coll2.find(u)
                for i in f:
                    coll2.delete_many(u)
                    lg.info('Record Deleted!')
                    continue
                raise Exception ('Record Deletion Error')
            except Exception as e:
                lg.exception(e)
  
```
